```python
import pandas as pd 
import requests
import json
```

```python
# df = pd.read_json('http://api.marketstack.com/v1/currencies?access_key=584524392ea6e2eb48f299c89421bf44')
headers =  {'content-type' : 'application/json',
                                       }
JSONContent  = requests.get('http://api.marketstack.com/v1/tickers?access_key=584524392ea6e2eb48f299c89421bf44',
                                      headers=headers, verify=True)
print(JSONContent.text)

            
```

```python
js = json.loads(JSONContent.text)
print(js)
```

```python
df = pd.json_normalize(js['data'])
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('.', '_').str.replace(',','_').str.replace('/','_').str.replace(':','').str.replace(':','')
```

```python
df['new_column'] = df.apply(lambda row: row.stock_exchange_acronym +row.stock_exchange_mic, axis = 1) 
```
